## Library Import
Below are the library we imported for loading and exploring the loan dataset

```python
from pymongo import MongoClient
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
client = MongoClient("mongodb://localhost:27017/")
collection = client["lendingclub_db"]["loans_raw"]
```

```python
df = pd.DataFrame(list(collection.find()))

# Drop the MongoDB internal _id column
df = df.drop(columns=["_id"])
df.head()
```

```python
df.info()
```

```python
df.dropna(how='all', axis=1, inplace=True)
df.info()
```

```python
df.drop(columns=['Unnamed: 0','id','member_id'], inplace=True)
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

```python
df.info()
```

```python
str_list=[]

for x in df:
    if df[x].dtype == 'object':
        str_list.append(x)
print(str_list)
```

```python
df[str_list].transpose().head(22)
```

```python
df[str_list].info()
```

```python
for col in str_list:
    if col=='issue_d' or col=='earliest_cr_line' or col=='last_pymnt_d' or col=='next_pymnt_d' or col=='last_credit_pull_d':
        str_list.remove(col)

print(str_list)
```

```python
str_list.remove('next_pymnt_d')
str_list.remove('term')
str_list.remove('emp_length')
```

```python
date_list=['issue_d','earliest_cr_line','last_pymnt_d','next_pymnt_d','last_credit_pull_d']
df[date_list].head()
```

```python
for c in date_list:
	df[c] = pd.to_datetime(df[c], format='%b-%y', errors='coerce')
df[date_list].head()
```

```python
df['term'] = (df['term'].str.replace(' months', '').astype(int))
df['emp_length'] = df['emp_length'].fillna('0').str.replace(' years', '').str.replace(' year', '').str.replace('< 1', '0').str.replace('10+', '10').astype(int)
df[['term','emp_length']].head()
```

### How to treat null values
- annual_inc: remove rows with null values
- delinq_2yrs: set to 0
- inq_last_6mths: set to 0
- mths_since_last_delinq: set to 0
- mths_since_last_record: set to 0
- open_acc: set to 1
- pub_rec: set to 0
- revol_util: set to 0
- total_acc: set to 1
- collections_12_mths_ex_med: set to 0
- mths_since_last_major_derog: set to 0
- acc_now_delinq: set to 0
- tot_coll_amt: set to 0
- tot_cur_bal: set to 0
- total_rev_hi_lim: set to 0

```python
df=df[df['annual_inc'].notna()]
df['open_acc'].fillna(1, inplace=True)
df['total_acc'].fillna(1, inplace=True)
```

```python
for col in ['delinq_2yrs', 'inq_last_6mths','mths_since_last_delinq','mths_since_last_record','pub_rec','revol_util','collections_12_mths_ex_med','mths_since_last_major_derog','acc_now_delinq','tot_coll_amt','tot_cur_bal','total_rev_hi_lim']:
    df[col].fillna(0, inplace=True)
```
